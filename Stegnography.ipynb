```python
#import libraries

import cv2 #opencv
import numpy as np
import string
import os
import matplotlib.pyplot as plt
```

```python
#ASCII CONVERSTION

d={chr(i):i for i in range(255)} #characeter to ascii
c={i:chr(i) for i in range(255)} #ascii to character
```

```python
image_path="/home/rgukt-basar/Downloads/spid.jpg"
x=cv2.imread(image_path)
xrbg = cv2.cvtColor(x,cv2.COLOR_BGR2RGB)
plt.imshow(xrbg)
plt.axis("off")
plt.show()
```

```python
x.shape
```

```python
key="123"
text = "secret"
```

```python
#ASCII Value

text_ascii=[d[ch]  for ch in text]  #list of ASCII values of text characters
key_ascii=[d[ch] for ch in key]    #list of ASCII values of key characters
print(text_ascii)
print(key_ascii)
```

```python
#encyrpt using pixel modification

x_enc=x.copy()
n=0  #no of rows
m=0  #no of col
z=0  #color panel
l=len(text)
kl=0
for i in range(l):
       orig_val=x_enc[n,m,z]
       new_val=d[text[i]]^d[key[kl]]
       x_enc[n,m,z]=new_val
       print(f"Embedding '{text[i]}' (ASCII {d[text[i]]}) XOR '{key[kl]}' (ASCII {d[key[kl]]}) = {new_val} at pixel ({n},{m},{z}) [original={orig_val}]") 
       n=n+1
       m=m+1
       m=(m+1)%3
       z=(z+1)%3
       kl=(kl+1)%len(key)
```

```python
cv2.imwrite("encrypt.jpg",x_enc)
```

```python
plt.imshow(cv2.cvtColor(x_enc,cv2.COLOR_BGR2RGB))
plt.title("Encrypted_Image")
plt.axis("off")
plt.show()
```

```python
#Decryption 
n,m,z=0,0,0
kl=0
decrypt=""
for i in range(l):
    val=x_enc[n,m,z]
    orig_char=c[val^d[key[kl]]]
    decrypt=decrypt+orig_char
    print(f"Decrypting pixel ({n},{m},{z}):{val} XOR {d[key[kl]]} = {val ^ d[key[kl]]} -> '{orig_char}'")
    n=n+1
    m=m+1
    m=(m+1)%3
    z=(z+1)%3
    kl=(kl+1)%len(key)
    
print(decrypt)
```
